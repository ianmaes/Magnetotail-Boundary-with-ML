```python
import pyspedas
from pytplot import tplot  
from pyspedas import get_data
import numpy as np
import matplotlib.pyplot as plt
```

```python
pyspedas.projects.omni.data(trange=['2010-11-5', '2024-11-6'],  datatype='hourly',)
```

```python
velocity = pyspedas.get_data('V')
sigma_velocity = pyspedas.get_data('SIGMA-V')
pressure = pyspedas.get_data('Pressure')
bz_gsm = pyspedas.get_data('BZ_GSM')
imf = pyspedas.get_data('ABS_B')
```

```python
velocity
```

```python
%matplotlib widget
velocity_averaged = np.convolve(velocity.y, np.ones(720)/720, mode='valid')
sigma_velocity_averaged = np.convolve(sigma_velocity.y, np.ones(720)/720, mode='valid')
pressure_averaged = np.convolve(pressure.y, np.ones(720)/720, mode='valid')
bz_gsm_averaged = np.convolve(bz_gsm.y, np.ones(720)/720, mode='valid')
imf_averaged = np.convolve(imf.y, np.ones(720)/720, mode='valid')
plt.subplots(5, 1, figsize=(10, 12))
plt.subplot(5, 1, 1)
plt.plot(velocity.times[719:].astype("datetime64[s]"), velocity_averaged, label='24-hour Averaged Velocity', color='orange')
plt.subplot(5, 1, 2)
plt.plot(sigma_velocity.times[719:].astype("datetime64[s]"), sigma_velocity_averaged, label='24-hour Averaged Sigma-Velocity', color='blue')
plt.subplot(5, 1, 3)
plt.plot(pressure.times[719:].astype("datetime64[s]"), pressure_averaged, label='24-hour Averaged Pressure', color='green')
plt.subplot(5, 1, 4)
plt.plot(bz_gsm.times[719:].astype("datetime64[s]"), bz_gsm_averaged, label='24-hour Averaged Bz_GSM', color='red')
plt.subplot(5, 1, 5)
plt.plot(imf.times[719:].astype("datetime64[s]"), imf_averaged, label='24-hour Averaged IMF', color='purple')
```
