#### **Demo script for data plotting.**

Data types:
- peef: electron full
- peer: electron reduced
- peeb: electron burst
- peif: ion full
- peir: ion reduced
- peib: ion burst

```python
import pyspedas
from pyspedas import tplot, options  
```

```python
trange = ['2010-02-13', '2010-02-14']
probe = 'b'
data = pyspedas.projects.themis.esa(trange=trange, probe=probe, downloadonly=True)
```

```python
data
```

```python
options('th'+probe+'_peef_en_eflux', 'ylog', True)
options('th'+probe+'_peef_en_eflux', 'zlog', True)

tplot(['th'+probe+'_peef_density', 'th'+probe+'_peef_flux', 'th'+probe+'_peef_en_eflux'])
```

```python
options('th'+probe+'_peif_en_eflux', 'ylog', True)
options('th'+probe+'_peif_en_eflux', 'zlog', True)

tplot(['th'+probe+'_peif_density', 'th'+probe+'_peif_velocity_dsl', 'th'+probe+'_peif_en_eflux'])
```

```python
options('th'+probe+'_peeb_en_eflux', 'ylog', True)
options('th'+probe+'_peeb_en_eflux', 'zlog', True)

tplot(['th'+probe+'_peeb_density', 'th'+probe+'_peeb_velocity_dsl', 'th'+probe+'_peeb_en_eflux'])
```
